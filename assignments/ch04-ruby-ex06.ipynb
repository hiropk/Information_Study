## 📖 学習内容：関数（メソッド）

同じ処理を複数箇所で使う場合に，抜き出して定義したものを**関数（Rubyではメソッド）**という。

```ruby
def メソッド名(引数名)
  処理
  return 戻り値
end
```

- **引数**：メソッドに渡す値
- **戻り値**：メソッドが返す値
- `return` で戻り値を返す
