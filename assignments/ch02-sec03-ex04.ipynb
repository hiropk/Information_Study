## ✅ 使い方
- `▶ ボタン` または `Shift+Enter` でセルを実行できる
- `html_code = """` で囲まれた部分が **index.html の中身そのもの**
- `<!-- コメント -->` を読んで，**コメントの下（または空いている場所）** に HTML を書こう
- 実行すると HTML コードと **プレビュー** が表示される

### ⚠️ 超重要：`<!-- コメント -->` は画面に出ない！

`<!-- ... -->` で囲んだ部分は **説明用メモ** であって，プレビューには表示されない。  
実際に画面に出すには，**コメントの外** に `<h2>` や `<button>` を書く必要がある。

```html
<!-- これは画面に出ない -->
<h2>これは画面に出る</h2>
```

### うまくいったかの見分け方
| 書いたもの | プレビューで確認できること |
|---|---|
| `<title>地元の名所紹介</title>` | タブのタイトル（コード表示で確認） |
| `<h2>地元の名所紹介</h2>` | ページ上部に **大きな見出し** が出る |
| `<button>…</button>` が4つ | **ボタンが4つ** 表示される |

## 📖 p.62　Webページと HTML

**Webページ** … Webブラウザで閲覧できるインターネット上の文書  
**Webサイト** … 複数の Webページで構成される意味のあるまとまり

**HTML** … Webページを作成するための記述言語。`<` と `>` で挟まれた **タグ** で文章構造を定義する（**マークアップ**）。

| 要素 | 内容 |
|---|---|
| `html` | HTML文書を示す |
| `title` | ウィンドウのタイトルバーに表示する |
| `head` | ヘッダを示す |
| `body` | ボディ（本文）を示す |
| `h1`〜`h6` | 見出し（数字が小さいほど大きい） |
| `p` | 段落を示す |

HTMLファイルは **2つの部分** で構成される。

- **head要素** … 文書自体の情報（メタデータ）
- **body要素** … Webブラウザに表示される部分

## 📖 p.62 の解説

**❶ ハイパーリンク**

複数の文書を相互に関連付け，結び付ける仕組みを **ハイパーテキスト** といい，テキスト間を結び付ける参照を **ハイパーリンク** という。

**❷ タグの書き方**

タグは **半角英数字** で記す。HTMLは大文字でも小文字でも区別しないが，小文字で記されることが多い。

**❸ 終了タグのないタグ**

改行を示す `<br>` のように，単独で機能し終了タグがないものもある。`<br/>` としてもよい。

**❹ 見出しタグ h1〜h6**

数字が小さいほど大きな見出しを表す。`h1` は Webページの主題を表すのに **一度だけ** 使うのが理想的。

## 📖 p.63　CSS と Webサイトの制作

**CSS（スタイルシート）** … 文字や画像の大きさ・色・配置・背景色など，視覚的なスタイルを定義する規格。

- HTML → 文書の **構造** を記述
- CSS → レイアウトや **デザイン** を定義

CSSの書式：`セレクタ { プロパティ: 値; }`

例：`h1 { color: red; }` → h1見出しの文字色を赤にする

CSSを読み込むには，HTMLの head 部に次のように書く。

```html
<link rel="stylesheet" href="example.css">
```

## 📖 p.63 の解説

**❶ メタデータ**

head要素に記述する。文字コードや Webページのタイトルなどを指定する。

```html

```

**❷ CSSの大文字・小文字**

プロパティと値のアルファベットは大文字でも小文字でも構わないが，**セレクタは大文字と小文字を区別する**ので注意する。

**❸ classセレクタ・idセレクタ**

HTMLファイル全体ではなく，**一部の HTMLタグだけ** にスタイルを適用できる。

## 🎯 例題4　トップページの作成

トップページ（`index.html`）に，名所紹介ページ A〜D へリンクするメニューを設置してみよう。  
ファイル名は `index.html` と `info1.html`〜`info4.html` とする。

### サイトマップ（図5）

```
トップページ（index.html）
  ├─ ページA（info1.html）
  ├─ ページB（info2.html）
  ├─ ページC（info3.html）
  └─ ページD（info4.html）
```

### 考え方

`button` タグでボタンを作り，各紹介ページに移るようにする。

```html
<button onclick="location.href='URL'">テキスト</button>
```

### アルゴリズム
1. `head` に文字コードとタイトルを書く
2. `body` に見出し（`h2`）を書く
3. ボタン4つを作り，`onclick` でリンク先を指定する

### ヒント（教科書 図6）
- タイトル・見出しは「地元の名所紹介」
- ボタンのテキストは「紹介ページA」〜「紹介ページD」

````python
from IPython.display import HTML, display

# index.html を作ろう！
# ↓ <!-- --> は説明用メモ（画面に出ない）。★の行を書き換えてね。
html_code = """
<html>
<head>
  <meta charset="UTF-8">
  <!-- ② title：タブのタイトル（画面には出ない） -->
  <title>地元の名所紹介</title>
</head>
<body>
  <!-- ③ 見出し：下の h2 の中身を書き換えよう → 画面に大きく出る -->
  <h2>地元の名所紹介</h2>

  <!-- ④ ボタン4つ：下の4行を書き換えよう → 画面にボタンが出る -->
  <button onclick="location.href='info1.html'">紹介ページA</button>
  <button onclick="location.href='info2.html'">紹介ページB</button>
  <button onclick="location.href='info3.html'">紹介ページC</button>
  <button onclick="location.href='info4.html'">紹介ページD</button>

</body>
</html>
"""

print("=== index.html ===")
print(html_code)
print("\n=== プレビュー ===")
display(HTML(html_code))
````

# 🌟 特別課題　「私の好きなもの」プレゼンを HTML/CSS で作ろう

**情報I ／ 特別課題（2章3節）**

先日 Classroom で作った **Google スライドのプレゼン** を，HTML/CSS で書き直してみよう！

### やること

1. **自分の Google スライド** を横に開いておく
2. **写真アップロード用セル** で，手元の画像をアップロードする
3. **HTML/CSS の型** を編集して，スライドの内容を **1枚ずつ** コピーする
4. 実行してプレビューで確認する

### Google スライドの型 ↔ HTML の対応

| Google スライド | HTML の `<section>` | 書く内容 |
|---|---|---|
| 1枚目：表紙 | `class="slide cover"` | タイトル・クラス・氏名・日付 |
| 2枚目：紹介するもの | `class="slide"` ① | 何を紹介する？／なぜ好き？ |
| 3枚目：魅力① | `class="slide"` ② | 魅力1つ ＋ 写真 |
| 4枚目：魅力② | `class="slide"` ③ | 魅力1つ ＋ 写真 |
| 5枚目：魅力③ | `class="slide"` ④ | 魅力1つ ＋ 写真 |
| 6枚目：まとめ | `class="slide"` ⑤ | 要約・いちばん伝えたいこと |

### デザインのルール（スライドと同じ）

- 背景 **白** ／ 文字 **黒** ＋ **テーマカラー** ＋ **強調色**
- **ゴシック体**（CSS では `sans-serif`）
- **1スライド1項目**（情報を詰め込みすぎない）
- 写真は **アップロード** して使う（URL は不要）

### 写真の入れ方

1. 下の **📷 写真アップロード** セルを実行する
2. 「ファイルを選択」から手元の写真（jpg / png など）を選ぶ
3. **選んだ順番** が 魅力① → 魅力② → 魅力③ に対応する
4. HTML セル実行時，`<!-- IMAGE_CHARM1 -->` などの位置に自動で表示される

````python
import base64

# 📷 魅力①②③用の写真をアップロードしよう
# → Google Colab で「ファイルを選択」ボタンが出る
# → 最大3枚まで。1枚目=魅力①、2枚目=魅力②、3枚目=魅力③

def to_data_url(filename, data):
    ext = filename.rsplit(".", 1)[-1].lower()
    mime = {"jpg": "jpeg", "jpeg": "jpeg", "png": "png", "gif": "gif", "webp": "webp"}.get(ext, "jpeg")
    b64 = base64.b64encode(data).decode()
    return f"data:image/{mime};base64,{b64}"

img_charm1 = ""
img_charm2 = ""
img_charm3 = ""

try:
    from google.colab import files

    print("📷 魅力①②③用の写真を選んでアップロードしてね")
    print("   ※ 選んだ順番 → 1枚目=魅力① / 2枚目=魅力② / 3枚目=魅力③")
    uploaded = files.upload()

    labels = ["魅力①", "魅力②", "魅力③"]
    urls = [img_charm1, img_charm2, img_charm3]
    for i, (fname, data) in enumerate(uploaded.items()):
        if i >= 3:
            break
        urls[i] = to_data_url(fname, data)
        print(f"  ✅ {labels[i]}：{fname}")
    img_charm1, img_charm2, img_charm3 = urls

    if not uploaded:
        print("  ⚠️ ファイルが選ばれなかったよ。もう一度実行して選んでね。")
except ImportError:
    print("⚠️ 写真アップロードは Google Colab で実行してね（ローカルでは使えない）")
````

````python
from IPython.display import HTML, display

# 写真アップロードセルをまだ実行していないときのため
try:
    img_charm1
except NameError:
    img_charm1 = img_charm2 = img_charm3 = ""
    print("⚠️ 先に 📷 写真アップロード セルを実行してね")

# 「私の好きなもの」プレゼンの HTML/CSS 型
# ↓ Google スライドを見ながら，<!-- コメント --> の指示どおり書き換えていこう
html_code = """
<html>
<head>
  <meta charset="UTF-8">
  <!-- ページタイトル：スライドの表紙タイトルと同じにしよう
       例：<title>私の好きなサッカー</title>
  -->
  <title>私の好きな○○</title>

  <style>
    /* ===== CSS（デザイン設定）=====
       ★ テーマカラーと強調色を自分のスライドに合わせて変えよう
       → 見出しの色（--theme-color）と強調部分の色（--accent-color）が変わる
    */
    :root {
      --theme-color: #2563eb;   /* ★テーマカラー（見出しなど） */
      --accent-color: #f97316;  /* ★強調色 */
    }

    body {
      margin: 0;
      font-family: sans-serif;  /* ゴシック体 */
      background: #ffffff;      /* 背景：白 */
      color: #000000;           /* 文字：黒 */
      line-height: 1.6;
    }

    /* 1枚のスライド＝1つの section */
    .slide {
      min-height: 80vh;
      padding: 48px 32px;
      border-bottom: 1px solid #e5e5e5;
    }

    /* 表紙スライド：中央揃え */
    .slide.cover {
      display: flex;
      flex-direction: column;
      justify-content: center;
      align-items: center;
      text-align: center;
    }

    .slide h1 {
      color: var(--theme-color);
      font-size: 2.4rem;
      margin-bottom: 24px;
    }

    .slide h2 {
      color: var(--theme-color);
      font-size: 1.8rem;
      margin-bottom: 20px;
    }

    .slide p {
      font-size: 1.1rem;
      margin: 8px 0;
    }

    .slide .info {
      font-size: 1rem;
      color: #555555;
    }

    .slide .accent {
      color: var(--accent-color);
      font-weight: bold;
    }

    .slide img {
      max-width: 100%;
      max-height: 280px;
      margin-top: 16px;
      border-radius: 8px;
    }
  </style>
</head>
<body>

  <!-- ==========================================
       【スライド1：表紙】
       Googleスライド1枚目と同じ内容を書こう
       → 実行すると中央に大きなタイトルが出る

       書き換える場所：
       ・<h1> の中 → 「私の好きな○○」
       ・<p class="info"> → クラス・番号・氏名
       ・<p class="info"> → 日付

       例：
       <h1>私の好きなサッカー</h1>
       <p class="info">2年3組 15番　山田太郎</p>
       <p class="info">2026年5月20日</p>
  ========================================== -->
  <section class="slide cover">
    <h1>私の好きな○○</h1>
    <p class="info">○年○組○番　氏名を入力</p>
    <p class="info">2026年　　月　　日</p>
  </section>

  <!-- ==========================================
       【スライド2：紹介するもの】（序論）
       Googleスライド2枚目と同じ内容を書こう
       → 見出し「紹介するもの」＋ 何を紹介するか ＋ なぜ好きか

       例：
       <h2>紹介するもの</h2>
       <p>・サッカーというスポーツ</p>
       <p>・<span class="accent">チームで勝つ喜び</span>が好きだから</p>
  ========================================== -->
  <section class="slide">
    <h2>紹介するもの</h2>
    <p>・何を紹介する？</p>
    <p>・なぜ好き？（ひとことで）</p>
  </section>

  <!-- ==========================================
       【スライド3：魅力①】
       Googleスライド3枚目と同じ内容を書こう
       → 魅力を1つ ＋ 写真（1スライド1項目！）
       → 写真は上のセルでアップロード済みなら <!-- IMAGE_CHARM1 --> に自動表示

       例：
       <h2>魅力①</h2>
       <p>・パスがつながったときの爽快感</p>
       <!-- IMAGE_CHARM1 -->
  ========================================== -->
  <section class="slide">
    <h2>魅力①</h2>
    <p>・魅力を1つ書こう</p>
    <!-- IMAGE_CHARM1 -->
  </section>

  <!-- ==========================================
       【スライド4：魅力②】
       Googleスライド4枚目と同じ内容を書こう
       → 写真は <!-- IMAGE_CHARM2 --> に自動表示
  ========================================== -->
  <section class="slide">
    <h2>魅力②</h2>
    <p>・魅力を1つ書こう</p>
    <!-- IMAGE_CHARM2 -->
  </section>

  <!-- ==========================================
       【スライド5：魅力③】
       Googleスライド5枚目と同じ内容を書こう
       → 写真は <!-- IMAGE_CHARM3 --> に自動表示
  ========================================== -->
  <section class="slide">
    <h2>魅力③</h2>
    <p>・魅力を1つ書こう</p>
    <!-- IMAGE_CHARM3 -->
  </section>

  <!-- ==========================================
       【スライド6：まとめ】（結論）
       Googleスライド6枚目と同じ内容を書こう
       → 魅力の要約 ＋ いちばん伝えたいこと

       例：
       <h2>まとめ</h2>
       <p>・チームワーク・爽快感・世界中に仲間がいる</p>
       <p>・<span class="accent">みんなにもサッカーをしてほしい！</span></p>
  ========================================== -->
  <section class="slide">
    <h2>まとめ</h2>
    <p>・魅力を要約しよう</p>
    <p>・いちばん伝えたいこと</p>
  </section>

</body>
</html>
"""

# アップロードした写真を HTML に埋め込む（上のセルを先に実行してね）
def make_img(data_url, alt):
    if data_url:
        return f'<img src="{data_url}" alt="{alt}">'
    return "<!-- 写真未アップロード（上のセルで選んでね） -->"

html_preview = html_code
html_preview = html_preview.replace("<!-- IMAGE_CHARM1 -->", make_img(img_charm1, "魅力①の写真"))
html_preview = html_preview.replace("<!-- IMAGE_CHARM2 -->", make_img(img_charm2, "魅力②の写真"))
html_preview = html_preview.replace("<!-- IMAGE_CHARM3 -->", make_img(img_charm3, "魅力③の写真"))

print("=== 私の好きなもの プレゼン（HTML/CSS）===")
print(html_preview)
print("\n=== プレビュー ===")
display(HTML(html_preview))
````
