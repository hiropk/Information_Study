## 🎯 問題

正しいパスワード（`Joho1`）が入力されるまで繰り返し入力を求めるプログラムを作成してみよう。

### アルゴリズム
1. 正しいパスワードを `a` に代入
2. 入力値を `b` に代入
3. `a` と `b` が等しくない間，繰り返し
4. 　次の入力値を `b` に代入
5. 正しく入力されたことを表示

### ヒント
- `!=` は「等しくない」
- `inputs.shift` で配列の先頭から値を取り出せる

````python
%%bash
ruby << 'RUBY'
a = "Joho1"
# ★変えよう★（試行するパスワードを順番に入れてみよう）
inputs = ["abc", "joho1", "Joho1"]

b = inputs.shift
puts "入力: #{b}"

# ③ aとbが等しくない間繰り返す
while # ここを入力
  b = inputs.shift
  puts "再入力: #{b}"
end

# ⑤ 正しく入力されたことを表示
# ここを入力

RUBY
````
