## 📖 学習内容：変数・基本演算

**変数**：データを格納する領域。Rubyでは型宣言不要で代入すると使える。

| 命令文 | 説明 |
|---|---|
| `puts 値` | 値を表示して改行 |
| `print 値` | 値を表示（改行なし） |

| 演算子 | 意味 | 演算子 | 意味 |
|---|---|---|---|
| `+` | 加算 | `-` | 減算 |
| `*` | 乗算 | `/` | 除算 |
| `%` | 余り | `**` | べき乗 |
