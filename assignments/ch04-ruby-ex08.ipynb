## 📖 学習内容：二分探索

**二分探索**：データを半分に分けて目的データを絞り込むアルゴリズム。
**あらかじめ整列されているデータ**に対して使える。

| 変数名 | 格納する値 |
|---|---|
| `imin` | 探索範囲の最小添字 |
| `imax` | 探索範囲の最大添字 |
| `imid` | 探索範囲の中央添字 |

## 🎯 問題

商品番号 `2001` を二分探索で探索するプログラムを作成してみよう。

### 探索の流れ（商品番号2001を探す場合）

| ステップ | 範囲 | 中央 | 中央の商品番号 | 次の範囲 |
|---|---|---|---|---|
| Ⅰ | 添字0〜5 | 添字2 | 1004（小さい） | 後半：添字3〜5 |
| Ⅱ | 添字3〜5 | 添字4 | 1011（小さい） | 後半：添字5〜5 |
| Ⅲ | 添字5〜5 | 添字5 | **2001（発見！）** | 終了 |

### ヒント
- 中央の添字：`(imin + imax) / 2`（整数どうしの割り算は整数になる）
- 中央より大きければ `imax = imid - 1`（前半を探す）
- 中央より小さければ `imin = imid + 1`（後半を探す）

```python
%%bash
ruby << 'RUBY'
stock = [
  [1001, "飲料水A",   87],
  [1003, "飲料水B",   89],
  [1004, "飲料水C",   52],
  [1007, "飲料水D",  120],
  [1011, "飲料水E",  119],
  [2001, "菓子パンA", 67]
]

# ★変えよう★
srch_stck = 2001

imin = 0  # 探索範囲の最小添字
imax = 5  # 探索範囲の最大添字

while imin <= imax
  # ⑤ 中央の添字を求める
  imid = # ここを入力

  if stock[imid][0] == srch_stck  # 発見！
    puts "#{stock[imid][0]} #{stock[imid][1]} #{stock[imid][2]}"
    break
  elsif # ここを入力（探索対象が前半にある条件）
    imax = # ここを入力
  else  # 後半にある
    imin = # ここを入力
  end
end

RUBY
```
