## PART 2-① 数式で確率を計算

$$P(\text{一致あり}) = 1 - \frac{365}{365} \times \frac{364}{365} \times \cdots \times \frac{365-n+1}{365}$$

````python
def birthday_probability(n, days=365):
    """n人で同誕生日ペアが少なくとも1組いる確率"""
    if n > days:
        return 1.0
    p_all_different = 1.0
    for i in range(n):
        p_all_different *= (days - i) / days
    return 1 - p_all_different


print(f"366人 → {birthday_probability(366)*100:.1f}%（鳩の巣原理で必ず一致）\n")
print(f"{'人数':>4}  {'一致確率':>10}")
print("-" * 18)
for n in [10, 20, 22, 23, 24, 30, 40, 50]:
    mark = " ← 50%超！" if n == 23 else ""
    print(f"{n:>4}  {birthday_probability(n)*100:>9.1f}%{mark}")
````

💡 **23人で約50.7%**。366人から一気に下がって，23人で50%ラインを超える。

---
## PART 2-② 乱数シミュレーションで実測

23人の誕生日をランダム生成して，「一致があるか」を **1万回** 試す。

````python
import random

random.seed(42)

PEOPLE = 23       # ★ 人数を変えてみよう
TRIALS = 10_000
DAYS = 365

match_count = sum(
    1 for _ in range(TRIALS)
    if len(set(random.randint(1, DAYS) for _ in range(PEOPLE))) < PEOPLE
)

sim_prob = match_count / TRIALS
theory_prob = birthday_probability(PEOPLE)

print(f"{PEOPLE}人 × {TRIALS:,}回シミュレーション")
print(f"  実測：{sim_prob*100:.1f}%")
print(f"  理論：{theory_prob*100:.1f}%")
````
