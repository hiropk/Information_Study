## 📖 学習内容：分岐構造（if文）

```ruby

```

| 演算子 | 意味 | 演算子 | 意味 |

## 🎯 問題

整数が，偶数か奇数，もしくはゼロか判定するプログラムを作成してみよう。

### アルゴリズム
1. 変数 `a` に数値を代入
2. もし `a` が 0 と等しければ → `'ゼロ'` と表示
3. そうでなくもし `a` を2で割った余りが0ならば → `'偶数'` と表示
4. そうでなければ → `'奇数'` と表示

### ヒント
- `if a == 0` でゼロ判定
- `a % 2 == 0` で偶数判定
- `elsif`（else if の短縮形）を使う

````python
%%bash
ruby << 'RUBY'
# ★変えよう★（0, 偶数, 奇数 の3パターンで試してみよう）
a = 7

# ② if文で判定
if # ここを入力
  puts "ゼロ"
elsif # ここを入力
  puts "偶数"
else
  # ここを入力
end

RUBY
````
