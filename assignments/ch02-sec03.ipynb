## 📋 必ず含めること（必須項目）

- 自分の名前・クラス・番号
- 自己紹介文（好きなこと・趣味・将来の夢など）
- 見出し・段落・箇条書きを使う
- CSSでスタイルをつける（色・フォント・余白など）

## 📊 採点ルーブリック（事前公開）

| 観点 | 20点 | 15点 | 10点 | 5点 |
|---|---|---|---|---|
| ① HTML構造 | 正しいタグが適切に使われている | ほぼ正しいが一部誤りがある | 基本タグはあるが構造が不完全 | タグの誤りが多い |
| ② コンテンツ | 必須情報がすべてあり内容が充実 | 必須情報はあるが内容が薄い | 必須情報の一部が欠けている | 情報がほとんどない |
| ③ CSSスタイル | 配色・フォント・余白が適切に設定されている | スタイルはあるが工夫が少ない | 一部スタイルがあるが不十分 | ほとんどスタイルがない |
| ④ 情報デザイン | 読みやすさ・配色・レイアウトが意識されている | ある程度意識されている | 一部意識されているが不十分 | ほとんど意識されていない |
| ⑤ 工夫・独自性 | 自分らしい工夫が随所に見られる | 工夫は見られるが全体的に平凡 | 少し工夫が見られる | ベースをほぼそのまま提出 |

**合計：5観点 × 20点 ＝ 100点満点**

## 📖 よく使うHTMLタグ・CSSプロパティ一覧

### HTMLタグ

| タグ | 意味 | 例 |
|---|---|---|
| `<h1>〜</h1>` | 大見出し | `<h1>自己紹介</h1>` |
| `<h2>〜</h2>` | 中見出し | `<h2>好きなこと</h2>` |
| `<p>〜</p>` | 段落（文章） | `<p>こんにちは！</p>` |
| `<ul>〜</ul>` | 箇条書きのまとまり | ― |
| `<li>〜</li>` | 箇条書きの1項目 | `<li>サッカー</li>` |
| `<section>〜</section>` | セクション（まとまり） | ― |
| `<strong>〜</strong>` | 太字・強調 | `<strong>重要</strong>` |

### CSSプロパティ

| プロパティ | 意味 | 例 |
|---|---|---|
| `color` | 文字色 | `color: #333333;` |
| `background-color` | 背景色 | `background-color: #f0f8ff;` |
| `font-size` | 文字サイズ | `font-size: 18px;` |
| `font-family` | フォント | `font-family: 'Noto Sans JP', sans-serif;` |
| `padding` | 内側の余白 | `padding: 16px;` |
| `margin` | 外側の余白 | `margin: 8px 0;` |
| `border-radius` | 角の丸み | `border-radius: 8px;` |
| `text-align` | 文字の揃え | `text-align: center;` |

### 色の指定方法

色は16進数（例：`#3a86ff`）や色名（例：`royalblue`）で指定できるよ。
→ [色を調べるサイト（Color Hunt）](https://colorhunt.co/) を参考にしてみよう！

## ✏️ ここを編集しよう！

下のコードセルの `<!-- ★ここを書き換えよう★ -->` のコメントを参考に，自分の情報に書き換えよう。

**セルを実行（▶ボタン または Shift+Enter）すると，すぐ下にプレビューが表示されるよ！**

```python
%%html
<!DOCTYPE html>
<html lang="ja">
<head>
  <meta charset="UTF-8">
  <title>自己紹介</title>
  <style>
    /* =====================
       ★CSSをここに書こう★
       全体のスタイルを設定するよ
    ===================== */
    body {
      font-family: 'Hiragino Sans', 'Meiryo', sans-serif;
      background-color: #f5f5f5; /* ★背景色を変えてみよう★ */
      color: #333333;           /* ★文字色を変えてみよう★ */
      max-width: 600px;
      margin: 0 auto;
      padding: 24px;
    }

    h1 {
      background-color: #3a86ff; /* ★テーマカラーを変えてみよう★ */
      color: white;
      padding: 16px 24px;
      border-radius: 8px;
      font-size: 28px;
    }

    h2 {
      color: #3a86ff;            /* ★h1と同じテーマカラーにそろえよう★ */
      border-left: 4px solid #3a86ff;
      padding-left: 10px;
      margin-top: 32px;
    }

    section {
      background-color: white;
      border-radius: 8px;
      padding: 16px 24px;
      margin-top: 16px;
    }

    ul {
      padding-left: 20px;
    }

    li {
      margin: 6px 0;
      line-height: 1.8;
    }

    p {
      line-height: 1.8;
    }

    /* ★自由にCSSを追加してみよう★ */

  </style>
</head>
<body>

  <!-- ★ページのタイトル（大見出し）★ -->
  <h1>自己紹介</h1>

  <!-- ★基本情報セクション★ -->
  <section>
    <h2>基本情報</h2>
    <ul>
      <li>名前：★ここに名前を書こう★</li>
      <li>クラス・番号：★例：25R・1番★</li>
    </ul>
  </section>

  <!-- ★自己紹介文セクション★ -->
  <section>
    <h2>自己紹介</h2>
    <p>★ここに自己紹介文を書こう。好きなことや得意なことを自由に書いてOK！★</p>
  </section>

  <!-- ★好きなこと・趣味セクション★ -->
  <section>
    <h2>好きなこと・趣味</h2>
    <ul>
      <li>★好きなこと1★</li>
      <li>★好きなこと2★</li>
      <li>★好きなこと3★</li>
    </ul>
  </section>

  <!-- ★将来の夢・目標セクション★ -->
  <section>
    <h2>将来の夢・目標</h2>
    <p>★ここに将来の夢や目標を書こう★</p>
  </section>

  <!-- ★自由に追加してもOK！セクションを増やしたり，デザインを工夫しよう★ -->

</body>
</html>
```

## 💡 デザインの工夫ヒント

授業で学んだ情報デザインの知識を活かしてみよう！

- **配色**：使う色は3〜4色に絞ろう（テーマカラー・背景色・文字色・強調色）
- **フォント**：読みやすさ重視のゴシック体がおすすめ
- **余白**：`padding` や `margin` をうまく使って，詰め込みすぎないようにしよう
- **コントラスト**：文字色と背景色の明度に差をつけると読みやすい
- **カラーUD**：赤と黒のような見分けにくい組み合わせは避けよう
- **視線誘導**：上から下・左から右の流れに沿ったレイアウトにしよう
