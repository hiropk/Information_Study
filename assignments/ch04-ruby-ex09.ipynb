## 📖 学習内容：バブルソート（整列）

**整列**：データをある値に基づいて並べ替えること。
**バブルソート**：隣り合う要素を比較し，順序が逆なら入れ替えることを繰り返す。
最も大きいデータが末尾に「浮き上がる」（バブル）ことから名付けられた。

| 変数名 | 格納する値 |
|---|---|
| `i` | 比較対象の添字 |
| `imax` | `i` の最大値 |
| `comp` | 計算量（比較回数） |

> **Rubyの多重代入**：`a, b = b, a` で一時変数不要で入れ替えができる！

## 🎯 問題

バラバラに格納された在庫データを，バブルソートで商品番号の昇順に整列させるプログラムを作成してみよう。

### バブルソートの流れ（4要素の場合）
**Ⅰ** 最大値を末尾（添字3）に移動
- 隣同士を比較 → 左が大きければ入れ替え → 最後まで繰り返す

**Ⅱ** 2番目に大きい値を添字2に移動（添字3はもう確定なので除く）

**Ⅲ** 3番目に大きい値を添字1に移動

### ヒント
- 外側のループ：`imax` を `last_ind-1` から `start_ind` まで1ずつ減らす
- 内側のループ：`i` を `start_ind` から `imax` まで繰り返す
- 順序が逆のとき：`stock[i], stock[i+1] = stock[i+1], stock[i]` で入れ替え
