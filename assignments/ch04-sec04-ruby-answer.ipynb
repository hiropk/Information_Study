# 1　プログラミングの基礎

## 1-1　変数を使用したプログラム

**変数**とは，データを格納する領域のことで，その名前を**変数名**という。

### Rubyの基本的な命令文

| 命令文 | 説明 |
|---|---|
| `puts 値` | 値を表示して改行する |
| `print 値` | 値を表示する（改行しない） |
| `gets.to_i` | ユーザの入力を整数型として受け取る |

### Rubyの基本的な演算子

|---|---|---|---|
| `+` | 加算 | `%` | 余り |
| `-` | 減算 | `/` | 除算（整数÷整数は整数部分のみ） |
| `*` | 乗算 | `**` | べき乗 |

## 1-3　反復構造を使用したプログラム

繰り返しの回数があらかじめわかる場合は `for` 文，わからない場合は `while` 文を使う。

```ruby

for 制御変数 in 初期値..終了値
```

## 1-4　配列

# 2　関数を使用したプログラム

## 2-1　関数（メソッド）とは

プログラムが大規模になると，同じ処理を複数の箇所で行うことが多くなる。
そこで，処理を抜き出して別に記述できる機能が**関数（Rubyではメソッド）**である。

関数は **引数** を受け取って処理を実行し，**戻り値** を返す。

```ruby
def メソッド名(引数名)
  処理
  return 戻り値
end
```

### 例題 5　自動販売機を関数と見立てたときの引数と戻り値

清涼飲料水の自動販売機も「引数を受け取って戻り値を返す関数」と見なせる。

| | 内容 |
|---|---|
| **引数** | 投入金額・押したボタン |
| **戻り値** | 清涼飲料水 |

**考察**：釣銭や売り切れのケースも戻り値として考えられる。単純に見える自動販売機も，実は複雑な関数である。

### 例題 6　四捨五入の処理をするユーザ定義関数

正の小数を小数第1位で四捨五入するメソッド `myround` を作成してみよう。

| | 名前 | 意味 |
|---|---|---|
| **引数** | `arg` | 四捨五入する値 |
| **戻り値** | | 四捨五入した結果 |

#### アルゴリズム
1. `arg` を引数としたメソッドを定義
2. `arg + 0.5` の値を `a` に代入
3. `a` の小数以下を切り捨てた値を `b` に代入
4. `b` を戻り値とする

````python
%%bash
ruby << 'RUBY'
# メソッドの定義
def myround(arg)
  a = arg + 0.5
  b = a.to_i
  return b
end

# 呼び出し側
# ★ここを変えよう★（いろいろな小数で試してみよう）
a = 7.4
puts myround(a)
b = 7.5
puts myround(b)
RUBY
````

> **参考　変数のスコープ**
> メソッドの内側で宣言した変数（**ローカル変数**）は，メソッドの外からはアクセスできない。
> 呼び出し側とメソッド内で同じ変数名 `a` を使っていても，別の変数として扱われる。

# 3　探索と整列のプログラム

## 3-1　線形探索のアルゴリズム

**探索**とは，たくさんのデータの中から目的のデータを探し出すことをいう。
最初のデータから順番に一つずつ調べていくアルゴリズムを**線形探索**という。

### 例題 7　線形探索で在庫数を調べるプログラム

下の在庫データから，商品番号を線形探索で探し，商品名と在庫数を表示するプログラムを作成してみよう。

| 変数名 | 格納する値 |
|---|---|
| `i` | 探索する添字 |
| `n` | データ数 |
| `srch_stck` | 探索対象の商品番号 |

````python
%%bash
ruby << 'RUBY'
stock = [
  [1001, "飲料水A",   87],
  [1003, "飲料水B",   89],
  [1004, "飲料水C",   52],
  [1007, "飲料水D",  120],
  [1011, "飲料水E",  119],
  [2001, "菓子パンA", 67]
]

# ★ここを変えよう★（探したい商品番号を入力）
srch_stck = 1007

n = stock.length  # データ数
for i in 0..n-1
  if srch_stck == stock[i][0]   # 商品が見つかった場合
    puts "#{stock[i][0]} #{stock[i][1]} #{stock[i][2]}"
    break  # 探索終了
  end
end
RUBY
````

> **考察**：線形探索は先頭から順番に調べるため，データ数が増えると繰り返し回数も増える。
> アルゴリズムは単純だが，大量データには不向きである。

---
## 3-2　二分探索のアルゴリズム

**二分探索**とは，データを半分に分けてどちらに目的データが含まれているか調べ，
さらに半分に分けることを繰り返すアルゴリズムである。
**あらかじめ整列されているデータ**に対して用いることができる。

### 例題 8　二分探索で在庫数を調べるプログラム

例題7の商品番号 `2001` を二分探索で探索するプログラムを作成してみよう。

#### 探索の流れ（商品番号2001を探す場合）

| ステップ | 範囲 | 中央 | 中央の商品番号 | 次の範囲 |
|---|---|---|---|---|
| Ⅰ | 添字0〜5 | 添字2 | 1004（小さい） | 後半：添字3〜5 |
| Ⅱ | 添字3〜5 | 添字4 | 1011（小さい） | 後半：添字5〜5 |
| Ⅲ | 添字5〜5 | 添字5 | **2001（発見！）** | 終了 |

````python
%%bash
ruby << 'RUBY'
stock = [
  [1001, "飲料水A",   87],
  [1003, "飲料水B",   89],
  [1004, "飲料水C",   52],
  [1007, "飲料水D",  120],
  [1011, "飲料水E",  119],
  [2001, "菓子パンA", 67]
]

# ★ここを変えよう★（探したい商品番号を入力）
srch_stck = 2001

imin = 0   # 探索範囲の最小添字
imax = 5   # 探索範囲の最大添字

while imin <= imax
  imid = (imin + imax) / 2          # 中央の添字
  if stock[imid][0] == srch_stck    # 発見！
    puts "#{stock[imid][0]} #{stock[imid][1]} #{stock[imid][2]}"
    break
  elsif stock[imid][0] > srch_stck  # 前半にある
    imax = imid - 1
  else                              # 後半にある
    imin = imid + 1
  end
end
RUBY
````

## 3-3　アルゴリズムの評価

同じ結果を得られるアルゴリズムが複数ある場合，**計算量**（処理の実行回数の目安）が少ないほうが優れているといえる。

| | 線形探索 | 二分探索 |
|---|---|---|
| データが2倍になると | 計算量も2倍 | 繰り返し1回分だけ増える |
| アルゴリズムの複雑さ | 単純 | やや複雑 |
| コードの長さ | 短い | やや長い |

→ 大量のデータを扱う場合は **二分探索のほうが優れている**。

## 3-4　整列のアルゴリズム（バブルソート）

データをある値に基づいて並べ替えることを**整列**という。
ここでは**バブルソート**と呼ばれる整列のアルゴリズムを学ぶ。

**考え方**：最も大きい商品番号のデータの場所から順に定めていく。
- 隣り合う2要素を比較し，順序が逆なら入れ替える
- これを繰り返すことで，大きいデータが末尾に「浮き上がる」（バブル）

### 例題 9　バブルソートを実行するプログラム

バラバラに格納された在庫データを，バブルソートで商品番号順に整列させるプログラムを作成してみよう。

| 変数名 | 格納する値 |
|---|---|
| `i` | 比較対象の添字 |
| `imax` | `i` の最大値 |
| `comp` | 計算量（比較回数） |

> **Rubyの多重代入による入れ替え**
> ```ruby
> a, b = b, a  # aとbの値を入れ替える（一時変数不要！）
> ```
> これを配列に使うと，行ごと一括で入れ替えられる。

````python
%%bash
ruby << 'RUBY'
stock = [
  [2006, "菓子パンA", 67],
  [2009, "菓子パンD", 82],
  [2010, "菓子パンE", 82],
  [1003, "飲料水C",   52]
]

comp      = 0  # 計算量
start_ind = 0  # 最初の要素の添字
last_ind  = 3  # 最後の要素の添字

imax = last_ind - 1
while imax >= start_ind     # 外側のループ
  for i in start_ind..imax  # 内側のループ（ネスト）
    comp += 1               # 計算量を加算
    if stock[i][0] > stock[i+1][0]  # 順序が逆なら入れ替え
      stock[i], stock[i+1] = stock[i+1], stock[i]  # 多重代入で1行！
    end
  end
  imax -= 1
end

for i in start_ind..last_ind  # 結果を表示
  puts stock[i].inspect
end
puts "計算量: #{comp}"
RUBY
````

> **考察**：Rubyの多重代入 `stock[i], stock[i+1] = stock[i+1], stock[i]` を使うと，
> 一時変数なしで行ごと一括入れ替えができる。
> Python版やマクロ版では複数行必要だった処理が，Rubyでは **1行** で書ける！

### ✏️ 確認問題 5

（1）例題9で，データ数を2倍・4倍に増やしたとき，計算量がどのように増加するか調べてみよう。

（2）例題9のプログラムを修正し，商品番号の**降順**に並べ替えてみよう。

````python
%%bash
ruby << 'RUBY'
# ★確認問題5(1)：データ数を増やして計算量を確認しよう★
stock = [
  [2006, "菓子パンA", 67],
  [2009, "菓子パンD", 82],
  [2010, "菓子パンE", 82],
  [1003, "飲料水C",   52]
  # ★ここにデータを追加してみよう★
]

comp      = 0
start_ind = 0
last_ind  = stock.length - 1

imax = last_ind - 1
while imax >= start_ind
  for i in start_ind..imax
    comp += 1
    if stock[i][0] > stock[i+1][0]  # ★(2)は ここの > を < に変えてみよう★
      stock[i], stock[i+1] = stock[i+1], stock[i]
    end
  end
  imax -= 1
end

for i in start_ind..last_ind
  puts stock[i].inspect
end
puts "計算量: #{comp}"
RUBY
````
