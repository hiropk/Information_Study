## PART 1：折ってみよう（手作業）

手元の紙を実際に折り続けてみよう。何回折れたか数えてメモしておいてね。

- 折るたびに厚さが **2倍** になる
- たぶん **7〜8回** で限界に…なぜ？

| 回数 | 厚さ |
|---|---|
| 1回 | 0.2 mm |
| 5回 | 3.2 mm |
| 7回 | 12.8 mm（1cm超！） |
| 8回 | 25.6 mm（2.5cm…） |

これが **2の指数関数（2ⁿ）** のパワー！

---
## PART 2：折り紙の厚さを計算しよう

### ステップ① まず数字で確認

下のセルを実行（▶ または Shift+Enter）してみよう！

```python
# 折り紙の厚さを計算しよう
# 折り紙の厚さ：約0.1mm = 0.0001m

thickness_m = 0.0001  # 最初の厚さ（メートル）

print(f"{'回数':>4}  {'厚さ（m）':>15}  {'わかりやすく'}")
print("-" * 50)

for n in range(0, 45):
    t = thickness_m * (2 ** n)
    if t < 1:
        desc = f"{t*1000:.4f} mm"
    elif t < 1000:
        desc = f"{t:.2f} m"
    elif t < 1_000_000:
        desc = f"{t/1000:.2f} km"
    else:
        desc = f"{t/1000:.0f} km"
    print(f"{n:>4}回  {t:>15.4f} m  {desc}")
```

### ステップ② グラフで見てみよう

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'IPAexGothic'

n_list = list(range(0, 45))
thickness_km = [0.0001 * (2**n) / 1000 for n in n_list]  # km換算

plt.figure(figsize=(10, 5))
plt.plot(n_list, thickness_km, color='royalblue', linewidth=2)
plt.axhline(y=384400, color='gray', linestyle='--', label='月までの距離（384,400 km）')
plt.xlabel('折った回数')
plt.ylabel('厚さ（km）')
plt.title('折り紙の厚さの変化（0〜44回）')
plt.yscale('log')  # 対数スケール
plt.legend()
plt.grid(True, which='both', alpha=0.3)
plt.tight_layout()
plt.show()

# 42回のときの厚さを確認
t_42 = 0.0001 * (2**42) / 1000
print(f"42回折ったときの厚さ：{t_42:,.0f} km")
print(f"月までの距離：384,400 km")
print(f"比較：{t_42/384400:.1f} 倍！")
```

## PART 3：自分のテーマで指数爆発を探求しよう

下のセルの ★ の部分を書き換えて実行してみよう！

**テーマ例**
- SNSで1人が3人に拡散 → 3人がさらに3人に…
- 毎日2倍のお小遣い → 30日後はいくら？
- ウイルスが2時間ごとに2倍に増殖 → 24時間後は？
- チェス盤の米粒（1マス1粒，2マス2粒…64マス目は？）

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'IPAexGothic'

# =============================================
# ★ここを書き換えよう★
# =============================================
title  = "★タイトルを入れよう★"   # グラフのタイトル
xlabel = "回数・日数・ステップ"      # 横軸のラベル
ylabel = "数・金額・量"              # 縦軸のラベル

start    = 1     # ★スタートの数（例：1人・1円・1個）★
multiply = 2     # ★何倍になるか（例：2倍・3倍）★
steps    = 30    # ★何回繰り返すか★
# =============================================

values = [start * (multiply ** n) for n in range(steps)]

plt.figure(figsize=(10, 5))
plt.plot(values, color='tomato', linewidth=2)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"最終ステップ（{steps-1}回後）の値：{values[-1]:,}")
```
