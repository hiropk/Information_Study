## 📖 学習内容：反復構造（for文）

繰り返しの回数があらかじめわかる場合は `for` 文を使う。

```ruby
for 制御変数 in 初期値..終了値
  処理
end
```

- `1..9` は1から**9を含む**範囲
- `1...9` は1から8まで（終了値を含まない）
