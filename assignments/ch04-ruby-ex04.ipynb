## 📖 学習内容：配列

**配列**：複数のデータをまとめて一つの名前で扱えるようにしたもの。
添字（インデックス）は **0** から始まる。

```ruby
a = []           # 空の配列を宣言
a.push(値)       # 末尾に要素を追加
a << 値          # push と同じ（Rubyらしい書き方）
puts a.inspect   # 配列の内容を一覧表示
a[2]             # 添字2の要素を取得
```

## ✏️ 確認問題
例題4のプログラムを修正して，添字1〜9の値だけを表示するプログラムを作成してみよう。（添字0 の値「0」は表示しない）
